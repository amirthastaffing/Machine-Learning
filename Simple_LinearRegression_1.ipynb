```python
import pandas as pd # importing  pandas library which has so many built-in classes and functions and assigned shortcut name as pd to call in future script
```

```python
dataset=pd.read_csv("Salary_Data.csv") #loading csv file from the database using pandas and assigning the csv file all data to the variable called 'dataset'
```

```python
dataset # viewing the data collection in the database
```

```python
independent= dataset[["YearsExperience"]] # assigning the input values from the dataset to input variable called independent 
dependent= dataset[["Salary"]] # assigning the output values from the dataset to output variable called dependent
```

```python
independent # viewing the input column
```

```python
dependent #displaying the output column
```

```python
from sklearn.model_selection import train_test_split  #importing sklearn library in the folder model selection and it contains many classes and functions. here we call train_test_split class.
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=.30,random_state=0)  # We call parameterized function train_test_split by passing 4 args and store the return values in 4 separate variables.
```

```python
x_train  #displaying train data set input
```

```python
x_test #displaying test dataset input
```

```python
from sklearn.linear_model import LinearRegression  #importing LinearRegression class from sklearn library
regressor=LinearRegression() # call the function LinearRegression and assing values of the function to the variable 'regressor'
Model_created=regressor.fit(x_train,y_train)    # using fit method passing the inputs(train dataset) and creating the model
```

```python
slope_weight=regressor.coef_     #finding the slope of the model. In what ratio the linear path is increasing?
```

```python
slope_weight
```

```python
bias=regressor.intercept_        # finding origin/initial value/starting value or to find the fresher's salary  
```

```python
bias
```

```python
y_predict=regressor.predict(x_test)          # Giving test dataset as input for a method called predict and predicting the value and assigning it to the variable
```

```python
y_predict    #displaying the predicted values. that is the output
```

```python
from sklearn.metrics import r2_score  # from the sklearn.metrics library calling the class or function r2_score
rvalue=r2_score(y_test,y_predict)     # validating the metrics by passing the actual value and predicted value to find the performance of the model
```

```python
rvalue    # since the value of r2 is nearing to 1 , it is considered as best model
```

```python
import pickle                     #importing the library to save the model
filename='Final_model_linear_regression_1.sav' # assigning the file path to the variable
pickle.dump(regressor,open(filename,'wb'))  #save the model using dump() method
```

```python
loaded_model = pickle.load(open('Final_model_linear_regression_1.sav','rb')) #Load the model to predict the value
result = loaded_model.predict([[13]])  # predicting the result for the given value.
print(result)
```
